```python
base_url='https://reelgood.com/curated/imdbs-best-rated-movies'
shows = []
for i in range(0,300, 50):
    soup=BeautifulSoup(requests.get(f"{base_url}?offset={i}").text, "html.parser")  
    shows += [s.get_text().replace(" ","%20") for s in soup.select("td.e126mwsw1 a[href^='/movie']")]
shows[0:5]
```

```python
results = []
dic = {}
base_url = 'https://reelgood.com' 
for q in shows: 
    src = f"{base_url}/search?q={q}"
    soup = BeautifulSoup(requests.get(f"{src}").text, "html.parser")  
    lnk = soup.select_one("div.e1qyeclq5 a")['href'] 
    try:    
        url=base_url+lnk   
        show = BeautifulSoup(requests.get(f"{url}").text, "html.parser")
        dic = {
            'query': q,
            'src': url, 
            'services': [s.get_text() for s in show.select("span.css-3g9tm3")] , 
        }
        results.append(dic)
        print(dic)
    except:
        print(f"scrape failed {url}")
df = pd.DataFrame(results)
df
```

```python
df.to_csv('data/top_shows.csv', index=False)
```
